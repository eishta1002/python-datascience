```python
df = df.rename(
    columns={
        'OdName':'country',
        'AreaName':'continent',
        'RegName':'region',
        'DevName':'status'
    }
)
```

```python
df.sort_values(by='total')
```

```python
df= df.sort_values(by='total', ascending=False)
```

```python
fig, ax =plt.subplots(figsize=(15,6))
sns.barplot(x='country' , y='total', data=df.head(10), ax=ax)
ax.axhline(df.head(10)['total'].mean(),color='r', linestyle='--')
ax.text(5, df.head(10)['total'].mean()+10000, df.head(10)['total'].mean())
plt.title('Top 10 countries with highest imigration overall')
plt.xticks(rotation=90)
plt.show()
```

grouping data 
-groupby(): group the data 
=groups the data according a single /multiple columns and then allows us to do the operations like mean,sum,maxetc
-pivot_table-pivot the data
=generate a matrix based on row and column provided

```python
df
```

```python
df.groupby(by='continent')
```

```python
#getting a particular group subset
df.groupby(by='continent').get_group('Asia')
```

```python
#getting group based stats
df.groupby(by='continent')['total'].sum()
```

```python

df.groupby(by='continent')[[1980,1981,1982,1983,1984]].sum(0)
```

```python

df.groupby(by='continent')[years].sum()
```

```python
df.groupby(by=['continent','status'])['total'].sum()
```

```python
df.groupby(by=['continent','status'])['total'].sum().reset_index()
```

```python
continent_df =df.groupby(by='continent')[years].sum()
sns.heatmap(continent_df, cmap ='coolwarm')
```

```python
cdf= df.groupby(by='continent')['total'].sum()
sns.barplot(x=cdf.index, y=cdf.values)
```

```python
fig , ax =plt.subplots(figsize =(15,6))
sns.lineplot(data=continent_df.T, dashes =False)
```

```python
#3items for pivot =index,column,values
#index and column should be categorical values and values must be numerical values
df2=df.pivot_table(index='continent',
               columns='status',
               values='total')
```

```python
df2
```

```python
sns.histplot(df2, kde=False)
```

```python
df2.plot(kind='pie',subplots=True, figsize=(15,6),autopct ='%1.1f%%')
```
